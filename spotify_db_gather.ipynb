```python
import config
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=config.client_id,
                                                           client_secret=config.client_secret))

```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
track= get_playlist_tracks("spotify", "7FnSjbyHaammTYOoGENfjD")
```

```python
song= track[4]["track"]["name"]
#artist= track[4]["track"]["artists"][0]["name"]
#uri= track[4]["track"]["uri"]
song
```

```python

songs= []
artists=[]
uris=[]
for i in range(len(track)):
    song= track[i]["track"]["name"]
    artist= track[i]["track"]["artists"][0]["name"]
    uri= track[i]["track"]["uri"]
    artists.append(artist)
    uris.append(uri)
    songs.append(song)
    
```

```python
df_song_artist= pd.DataFrame({"songs":songs,"artists":artists,"uris":uris})
df_song_artist
```
