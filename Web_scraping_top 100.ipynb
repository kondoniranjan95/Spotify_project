```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
url = "https://www.billboard.com/charts/hot-100/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
```

```python

x= soup.select(".c-title")
len(x)
```

```python
title= []
```

```python
num_iter= len(x)
for i in range(num_iter):
    title.append(soup.select(".c-title")[i].get_text(strip=True))
```

```python
title= title[4:104]
title
```

```python
y= soup.select("span.c-label")
len(y)
```

```python
artist=[]
num_iter= len(y)
for i in range(num_iter):
    artist.append(y[i].get_text(strip=True))
```

```python
artist = [item for item in artist if not item.isdigit()]
artist
```

```python
a= "NEW"
b= "-"
c= "RE-\nENTER"
for elem in list(artist): 
    if elem== a   or elem==b or elem==c: 
        artist.remove(elem)
```

```python
title
```

```python
df= pd.DataFrame({"title":title,"artist":artist})
df
```
